```python
# see the different types of regents exam with its mean score over time by grouping data
df.groupby(['Year','Regents Exam'])['Mean Score'].mean()
```

```python
# see the different types of school level with its mean score over time by grouping data
df.groupby(["Year", "School Level"])["Mean Score"].mean()
```

```python
# select English to narrow down scope
df["Regents Exam"].unique()
regents_name = "English"
regents_df = df[df["Regents Exam"] == regents_name]
```

```python
# filtering to look at mean score and year for English regents exams
regents_df_mean = regents_df[["Mean Score", "Year", "School Level"]]
```

```python
#drop rows with NaN
regents_df_mean_filtered = regents_df_mean.dropna(subset="Mean Score")
```

```python
#not the kind of graph I was looking for
regents_df_mean_filtered.plot(
    kind="line",
    x="Year", y="Mean Score",
    title = "Mean scores over time",
    legend = "School Level"
)
```

```python
#using groupby to summarize data then .unstack() function to use for plotting data
plot.data = regents_df_mean_filtered.groupby(["Year", "School Level"])["Mean Score"].mean().unstack()
```

```python
plot.data.plot(
    kind="line",
    xlabel="Year",
    ylabel="Mean Score",
    title="Mean Score Over Time by School Level"
)
```

Conclusion:
The mean score for the English Regents Exam decreased from 2015-2016 for secondary school, high school, and K-12 all grades school levels. High school students have a lower mean score compared to K-12 all grades school level.
